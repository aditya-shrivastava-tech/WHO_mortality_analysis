```python
# Read the specific sheet from the Excel file
country_excel_sheet = pd.read_excel('Disease codes.xlsx', sheet_name='country_codes')
# Creating a dictionary {code:country_name}
country_dict = country_excel_sheet.set_index('country')['name'].to_dict()
print(country_dict)
```
